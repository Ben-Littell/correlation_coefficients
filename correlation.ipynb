```python
import matplotlib.pyplot as plt
import math


x = [43, 21, 25, 42, 57, 59]
y = [99, 65, 79, 75, 87, 81]

def plot_data(dx, dy, pltcode):
    plt.plot(dx, dy, pltcode)
    plt.show()


plot_data(x, y, 'xr')
```

The following cell has a function to create find the correlation coefficient of the data with helper functions as needed. 

```python
def corcoeff(xd, yd):
    sigma1 = sigma_xy(xd, yd) * len(xd)
    sigma2 = sum(xd)*sum(yd)
    sigma3 = len(xd) * sum([val**2 for val in xd])
    sigma4 = sum(xd)**2
    sigma5 = len(yd) * sum([val**2 for val in yd])
    sigma6 = sum(yd)**2
    top = (sigma1 - sigma2)
    bottom = (math.sqrt(sigma3-sigma4))*(math.sqrt(sigma5-sigma6))
    return top/bottom
    

def sigma_xy(xd, yd):
    nlist = []
    for i in range(len(xd)):
        nlist.append((xd[i]*yd[i]))
    return sum(nlist)

    
    

correlation_xy = corcoeff(x, y)
print(correlation_xy)
# np.linalg.inv()
```

```python
x2 = [2, 3, 5, 7, 9]
y2 = [4, 5, 7, 10, 15]


def least_sqrs(xd, yd):
    matrix1 = [[sum(val**2 for val in xd), sum(xd)], [sum(xd), len(xd)]]
    matrix2 = [sigma_xy(xd, yd), sum(yd)]
    array1 = np.array(matrix1)
    array2 = np.array(matrix2)
    invarray1 = np.linalg.inv(array1)
    solution = np.dot(invarray1, array2)
    return solution


    
data2 = least_sqrs(x2, y2)
```

```python
def scatter_plot(data1, data2, slope, y_int):
    y_vals = []
    x_data = [min(data1), max(data1)]
    for val in range(2):
        ans = (slope * x_data[val]) + y_int
        y_vals.append(ans)
    plt.plot(x_data, y_vals, '-r')
    plt.scatter(data1, data2)
    plt.title('SCATTER PLOT')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.text(x_data[1], y_vals[1], f'Y={round(slope)}*X+{round(y_int)}', color='g')
    plt.show()
    
scatter_plot(x2, y2, data2[0], data2[1])
```
